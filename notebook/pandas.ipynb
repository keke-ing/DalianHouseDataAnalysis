```python
df.loc[:, "gross_area"] = df["gross_area"].str.strip("㎡")
```

```python
df["floor_number"] = df["floor_number"].str.extract('共(\d+)层', expand=False)
```

```python
df[['bedroom', 'living_room', 'kitchen', 'toilet']] = df["household_style"].str.extract('(\d+)室(\d+)厅(\d+)厨(\d+)卫',
                                                                                            expand=False)
```
